# How runnables work example

```python
import random

from pexpect.replwrap import python


class PseudoLLM:
    
    def __init__(self):
        print("LLM created")

    def predict(self, prompt):
        response_list = [
            "Delhi is the capital of india",
            "AI stands for artificial Intelligence",
            "IPL is Indian cricket league"
        ]

        return {"response": random.choice(response_list)}

llm = PseudoLLM()
llm.predict("what")
```

```python
class PseudoPromptTemplate:
    def __init__(self,template,input_variables):
        self.template = template
        self.input_variables = input_variables
    
    def format(self, input_dict):
        return self.template.format(**input_dict) # inbuilt str format function 

template = PseudoPromptTemplate(
    template="Write a {length} poem about {topic}",
    input_variables=["topic","length"] # just for demo of actual Prompt template, no use in this code
)

template.format({"topic":"India", "length":"short"})
```

```python
# just for understanding
"write a poem about {topic}".format(topic="India")
# ** unpacks to {"topic":"India"} to topic="India" 
```

# User using

```python
prompt = template.format({"length":"short","topic":"india"})

llm = PseudoLLM()
llm.predict(prompt)
```
