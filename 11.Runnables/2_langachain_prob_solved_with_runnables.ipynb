# Problem solved with runnables

## runnables is an abstract class

so everything has a common structure

```python
from abc import ABC, abstractmethod

class Runnable(ABC):
    
    @abstractmethod
    def invoke(self, input_data):
        pass
    
```

```python
import random

class PseudoLLM(Runnable):
    
    def __init__(self):
        print("LLM created")
    
    def invoke(self, input_data):
        response_list = [
            "Delhi is the capital of india",
            "AI stands for artificial Intelligence",
            "IPL is Indian cricket league"
        ]

        return {"response": random.choice(response_list)}



    def predict(self, prompt):
        print("This will be depretiated in the future use invoke()") # add warning 
        response_list = [
            "Delhi is the capital of india",
            "AI stands for artificial Intelligence",
            "IPL is Indian cricket league"
        ]

        return {"response": random.choice(response_list)}
```

```python
llm = PseudoLLM()
```

```python
class PseudoStrOutputParser(Runnable):
    def __init__(self):
        pass
    
    def invoke(self, input_data):
        return input_data["response"]
```

```python

class PseudoPromptTemplate(Runnable):
    def __init__(self,template, input_variables):
        self.template = template
        self.input_variables = input_variables
    
    def invoke(self, input_dict):
        return self.template.format(**input_dict) # inbuilt str format function 
    
    def format(self, input_dict):
        print("This will be depretiated in the future use invoke()") # add warning
        return self.template.format(**input_dict) # inbuilt str format function 
```

# Now use this

```python
class RunnableConnector(Runnable):
    
    def __init__(self, runnable_list:list):
        self.runnable_list = runnable_list

    def invoke(self, input_data):
        for runnable in self.runnable_list: #This loop would invoke each invoke in each class
            input_data = runnable.invoke(input_data) #? look below cell first the input data is prompt and in second loop the runnable is llm
                                                    #* and output of the previous step becomes input of current step
                                                    #! variable name is input_data and inside invoke(input_data)
                                                    #// This is how we form chains
        return input_data
```

```python
template = PseudoPromptTemplate(
    template="Write a {length} poem about {topic}",
    input_variables=["length", "topic"]
)

llm = PseudoLLM()
parser = PseudoStrOutputParser()

chain = RunnableConnector([template, llm, parser])
chain.invoke({"length":"long", "topic":"india"})
```

### Above we can see we connect `template`, `llm` and `parser` and just use `invoke()` only once
