RunnablePassthrough is a special primitive that simply returns the input as output without modifying it.

```python
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableParallel, RunnablePassthrough
from dotenv import load_dotenv

load_dotenv()

tweet_prompt = PromptTemplate.from_template(
    "Generate a short joke about {topic}"
)

explanation_prompt = PromptTemplate.from_template(
    "Explain this joke: {joke}"
)

model = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0
)

parser = StrOutputParser()

# Generate joke
joke_chain = tweet_prompt | model | parser

# Final parallel processing
final_chain = (
    joke_chain
    | RunnableParallel({
        "joke": RunnablePassthrough(),
        "explanation": explanation_prompt | model | parser
    })
)

result = final_chain.invoke({"topic": "cricket"})

print(result)
```
