RunnableBranch

is a control flow component in langchain that allows you to conditionally route input data to different chains or runnables based on custom logic

it functions like if/elif/else block for chains 

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableBranch, RunnablePassthrough

load_dotenv()

model = ChatGoogleGenerativeAI(
    model="models/gemini-2.5-flash-lite"
)
parser = StrOutputParser()

prompt1 = PromptTemplate.from_template(
    template="Write a detailed report on {topic}"
)

prompt2 =  PromptTemplate.from_template(
    template="Summarize the following text \n {text}"
)

parser = StrOutputParser()

report_gen_chain = prompt1 | model | parser

branch_chain = RunnableBranch(
    # (condition, runnable ),
    (lambda x: len(x.split()) > 500, prompt2 | model | parser),
    RunnablePassthrough() # default or else statement
)

final_chain = report_gen_chain | branch_chain

result = final_chain.invoke({"topic":"Russia Vs Ukraine"})
print(result)
```
