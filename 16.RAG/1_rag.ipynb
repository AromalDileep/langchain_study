## RAG
- Indexing    → Build knowledge base
- Retrieval  → Fetch relevant information
- Augmentation → Add retrieved context to prompt
- Generation → LLM generates answer

### Indexing Phase (Detailed Steps)

```text
1. Load Documents
2. Split Documents into Chunks
3. Convert Chunks to Embeddings
4. Store Embeddings in Vector Database
```
### Example Flow

```text
PDF / TXT / Website
        ↓
1. Document Loader
        ↓
Raw Documents
        ↓
2. Text Splitter
        ↓
Chunks
        ↓
3. Embedding Model
        ↓
Vectors
        ↓
4. Vector Database (Chroma/FAISS/Pinecone)
```
